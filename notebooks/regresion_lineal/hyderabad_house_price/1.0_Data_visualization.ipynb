# Hyderabad House Price Data Visualization

```python
import pandas as pd

house_prices_ds = pd.read_csv('../../../datasets/raw/hyderabad_house_price_with_nullables.csv')

samples = house_prices_ds.shape[0]
dimensions = house_prices_ds.shape[1]
numeric_cols = house_prices_ds.select_dtypes(include=["int64", "float64"]).columns.tolist()
categorical_cols = house_prices_ds.select_dtypes(include=["object", "category"]).columns.tolist()

print(f"Samples: {samples}")
print(f"Dimensions (Label + features): {dimensions}")
print(f"Categoricals: {len(categorical_cols)}")
print(f"Numerics: {len(numeric_cols)}")
print("Numeric features: ", numeric_cols)
print("Categoricals features: ", categorical_cols)

house_prices_ds.info()
```

## Outiers values (percetiles)

```python
import seaborn as sns
import matplotlib.pyplot as plt


for col in house_prices_ds.select_dtypes(include=["int64", "float64"]).columns:
    sns.boxplot(x=house_prices_ds[col])
    plt.title(f"Boxplot de {col}")
    plt.show()
```

## Distribution of individual variables

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Seleccionar columnas numéricas
numeric_cols = house_prices_ds.select_dtypes(include=["int64","float64"]).columns

# Definir layout de subplots
n_cols = 3  # cantidad de columnas de subplots
n_rows = int(np.ceil(len(numeric_cols) / n_cols))

fig, axes = plt.subplots(n_rows, n_cols, figsize=(5*n_cols, 4*n_rows))
axes = axes.flatten()  # lo convertimos a 1D para iterar fácil

# Crear histogramas en cada subplot
for i, col in enumerate(numeric_cols):
    sns.histplot(data=house_prices_ds, x=col, bins=10, ax=axes[i], kde=True)
    axes[i].set_title(f"Distribución de {col}")

# Eliminar ejes vacíos si sobran
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

## Correlation Matrix

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Remove the column location
house_prices_ds.drop(columns=["Location"])

# Find correlation between variables
# La correlación es una medida estadística que indica qué tan fuerte y en qué dirección se relacionan dos variables.
# Si aumenta una variable y también aumenta la otra, decimos que hay correlación positiva.
# Si aumenta una variable y la otra disminuye, hablamos de correlación negativa.
# Si no hay un patrón claro, la correlación es cercana a 0 (sin relación lineal).
numeric_cols = ["Price", "Area", "No. of Bedrooms", "Resale", "SwimmingPool", "Gymnasium"]
corr = house_prices_ds[numeric_cols].corr()
print(corr)

# HitMap
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de Correlación")
plt.show()
```
